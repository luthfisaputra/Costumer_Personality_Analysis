```python
import pandas as pd

df = pd.read_csv('/content/marketing_campaign.csv', sep='\t')

display(df.head())
```

```python
df.info()
```

```python
df.describe()
```

```python
# cek missing values
df.isnull().sum()
```

```python
# isi missing values di kolom Income dengan median
df['Income'] = df['Income'].fillna(df['Income'].mean())
```

```python
# cek dan hapus duplikat
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
# menambahkan kolom Age
df['Age'] = 2025 - df['Year_Birth']
df.head()
```

```python
# cek dan hapus outlier age > 80th
df[df['Age'] > 80]
df = df[df['Age'] < 80]
```

# **Descriptive Analytics**

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Distribusi rata rata umur pelanggan
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribusi umur pelanggan')
plt.xlabel('Umur')
plt.ylabel('jumlah')
plt.show()
```

```python
# Distribusi pendapatan pelanggan

sns.histplot(df['Income'], bins=50, kde=True)
plt.title('Distribusi pendapatan pelanggan')
plt.xlabel('Pendapatan')
plt.ylabel('jumlah')
plt.show()
```

```python
# Status pernikahan yang paling banyak
df['Marital_Status'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(6,6))
plt.title('Status pernikahan yang paling banyak')
plt.ylabel('')
plt.show()
```

```python
df['children'] = df['Kidhome'] + df['Teenhome']
avg_cildren = df['children'].mean()

plt.figure(figsize=(6,4))
df['children'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribusi jumlah anak per keluarga')
plt.xlabel('Jumlah keluarga')
plt.ylabel('Jumlah keluarga')
plt.show()
```

```python
produk = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
total_per_produk = df[produk].sum().sort_values(ascending=False)

total_per_produk.plot(kind='bar', color='skyblue')
plt.title('Total pembelian per produk')
plt.ylabel('Jumlah pembelian')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# **Descriptive Analytics**

```python
produk = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df['Total_Pengeluaran'] = df[produk].sum(axis=1)

bins = [0, 20000, 50000, 80000, 120000]
labels = ['low', 'medium', 'high', 'very high']
df['Income_group'] = pd.cut(df['Income'], bins=bins, labels=labels)

df.groupby('Income_group')['Total_Pengeluaran'].mean()

sns.barplot(x='Income_group', y='Total_Pengeluaran', data=df, estimator='mean', ci=None)
plt.title('Rata-rata pengeluaran berdasarkan tingkat pendapatan')
plt.xlabel('Kelompok pendapatan')
plt.ylabel('total pengeluaran')
plt.show()
```

```python
df['Jumlah_Anak'] = df['Kidhome'] + df['Teenhome']
produk = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df['Total_Pengeluaran'] = df[produk].sum(axis=1)

df[['Jumlah_Anak', 'Total_Pengeluaran']].corr()

sns.scatterplot(x='Jumlah_Anak', y='Total_Pengeluaran', data=df)
plt.title('Korelasi antara jumlah anak dan total pengeluaran')
plt.xlabel('Jumlah Anak')
plt.ylabel('Total Pengeluaran')
plt.show()

df.groupby('Jumlah_Anak')['Total_Pengeluaran'].mean()
```

```python
df[['Age', 'Response']].corr()

sns.boxplot(x='Response', y='Age', data=df)
plt.title("Distribusi Usia Berdasarkan Respon Campaign")
plt.xlabel("Respon Campaign")
plt.ylabel("Usia")
plt.show()

df.groupby('Response')['Age'].mean().plot()
plt.title('Rata-rata respon campaign berdasarkan usia')
plt.xlabel('Usia')
plt.ylabel('Respon campaign')
plt.show()

bins = [18, 30, 45, 60, 80]
labels = ['Muda', 'Dewasa', 'Paruh Baya', 'Lansia']
df['Kelompok_Usia'] = pd.cut(df['Age'], bins=bins, labels=labels)

df.groupby('Kelompok_Usia')['Response'].mean()
```

```python
produk = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df['Total_Pengeluaran'] = df[produk].sum(axis=1)

df.groupby('Marital_Status')['Total_Pengeluaran'].mean().sort_values(ascending=False)

sns.barplot(x='Marital_Status', y='Total_Pengeluaran', data=df, estimator='mean', ci=None)
plt.title("Rata-rata Pengeluaran Berdasarkan Status Pernikahan")
plt.xlabel("Status Pernikahan")
plt.ylabel("Total Pengeluaran")
plt.xticks(rotation=45)
plt.show()

```

```python
df[['NumWebPurchases', 'Response']].corr()

sns.boxplot(x='Response', y='NumWebPurchases', data=df)
plt.title("Jumlah Pembelian Online vs Respon Campaign")
plt.xlabel("Respon Campaign")
plt.ylabel("Jumlah Pembelian Online")
plt.show()

bins = [0, 2, 5, 10, 20]
labels = ['Jarang', 'Cukup', 'Sering', 'Sangat Sering']
df['WebPurchase_Level'] = pd.cut(df['NumWebPurchases'], bins=bins, labels=labels)

df.groupby('WebPurchase_Level')['Response'].mean()
```
